# Data Visualisation Using Seaborn

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import numpy as np
```

```python
# getting the built in dataset of seaborn
print(len(sns.get_dataset_names()))
sns.get_dataset_names()
```

```python
df= sns.load_dataset("iris")
```

```python
df
```

```python
df.species.unique()
```

```python
df.species.value_counts()
```

```python
df.head()
```

```python
import warnings
warnings.filterwarnings('ignore')
```

```python
sns.lineplot(x='sepal_length', y='sepal_width',data=df,ci=None)
```

```python
df
```

```python
# linee plot
sns.lineplot(x="sepal_length",y="sepal_width",data=df,hue='species',ci=None)
plt.show()
```

```python
# continous and categorical feature
sns.lineplot(x="sepal_length",y="species",data=df,hue='species',ci=None)
plt.show()
```

```python
sns.lineplot(x="petal_length",y="species",data=df,hue='species',ci=None)
plt.show()
```

```python
sns.scatterplot(x="sepal_length",y="sepal_width",data=df,hue='species')
plt.title("My ")
plt.show()
```

```python
df.head()
```

```python
sns.scatterplot(x="petal_length",y="petal_width",data=df,hue='species')
plt.title("Iris ScatterPlot")
plt.show()
```

```python
### Bar Plot  (categorical estimated plot)
# categorical and numerical data

sns.barplot(x='species',y='sepal_length', data=df)
plt.show()
```

```python
sns.barplot(x='species',y='petal_length', data=df)
plt.show()
```

```python
# Count plot #when u have only categorical data feature & # descreate data 

sns.countplot(x='species',data=df)
plt.show()
```

```python
# box plot # IQR representation
#the dots are indicating outlier
sns.boxplot(x='sepal_width', data=df)
plt.show()
```

```python
sns.boxplot(x='species', y='sepal_width', data=df)
plt.show()
```

```python
sns.boxplot(x='sepal_length', data=df)
plt.show()
```

```python
# histplot / histogram
sns.histplot(x='petal_width', data=df)
plt.show()
```

```python
sns.histplot(x='petal_width', data=df, hue='species',kde=True)
plt.show()
```

```python
import warnings
warnings.filterwarnings('ignore')
#distplot

sns.distplot(df['petal_width'],color='red') # hist=false, #kde=False
plt.show()
```

```python
# pair plot
import warnings
warnings.filterwarnings('ignore')

sns.pairplot(data=df, hue='species')
plt.show()
```

```python
# Heat map

tc=df.corr()
tc
```

```python
sns.heatmap(tc,annot=True)
plt.show()
```
